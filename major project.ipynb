```python
import tkinter as tk
from tkinter import *

class Button(tk.Button):
    def __init__(self, master, **kw):
        tk.Button.__init__(self,master=master,**kw)
        self.defaultBackground = self["background"]
        self.bind("<Enter>", self.on_enter)
        self.bind("<Leave>", self.on_leave)

    def on_enter(self, e):
        self['background'] = 'grey'

    def on_leave(self, e):
        self['background'] = self.defaultBackground

expression = ""
 

def press(num):

    global expression
 
    
    expression = expression + str(num)
 

    equation.set(expression)
 
 
def equalpress():

    try:
 
        global expression
 
        
        total = str(eval(expression))
 
        equation.set(total)
 
        
        expression = total
 

    except:
 
        equation.set(" error ")
        expression = total
 
 

def clear():
    global expression
    expression = ""
    equation.set("")
def clearsingle():
    global expression
    expression = expression[:len(expression)-1]
    equation.set(expression)


if __name__ == "__main__":
    gui = Tk()
 
    gui.configure(background="grey")
 
    gui.title("Calculator")
 
    gui.geometry("363x390")
 

    equation = StringVar()
 

    expression_field = Entry(gui, textvariable=equation,bg='grey',fg='white',borderwidth=0,font=('calbri 20 bold'))
 

    expression_field.grid(columnspan=4,ipady=20)
 

    button1 = Button(gui, text=' 7 ', fg='white', bg='black',
                    command=lambda: press(7), height=2, width=8,font=('calbri 12 bold'))
    button1.grid(row=3, column=0)
 
    button2 = Button(gui, text=' 8 ', fg='white', bg='black',
                    command=lambda: press(8), height=2, width=8,font=('calbri 12 bold'))
    button2.grid(row=3, column=1)
 
    button3 = Button(gui, text=' 9 ', fg='white', bg='black',
                    command=lambda: press(9), height=2, width=8,font=('calbri 12 bold'))
    button3.grid(row=3, column=2)
 
    button4 = Button(gui, text=' 4 ', fg='white', bg='black',
                    command=lambda: press(4), height=2, width=8,font=('calbri 12 bold'))
    button4.grid(row=4, column=0)
 
    button5 = Button(gui, text=' 5 ', fg='white', bg='black',
                    command=lambda: press(5), height=2, width=8,font=('calbri 12 bold'))
    button5.grid(row=4, column=1)
 
    button6 = Button(gui, text=' 6 ', fg='white', bg='black',
                    command=lambda: press(6), height=2, width=8,font=('calbri 12 bold'))
    button6.grid(row=4, column=2)
 
    button7 = Button(gui, text=' 1 ', fg='white', bg='black',
                    command=lambda: press(1), height=2, width=8,font=('calbri 12 bold'))
    button7.grid(row=5, column=0)
 
    button8 = Button(gui, text=' 2 ', fg='white', bg='black',
                    command=lambda: press(2), height=2, width=8,font=('calbri 12 bold'))
    button8.grid(row=5, column=1)
 
    button9 = Button(gui, text=' 3 ', fg='white', bg='black',
                    command=lambda: press(3), height=2, width=8,font=('calbri 12 bold'))
    button9.grid(row=5, column=2)
 
    button0 = Button(gui, text=' 0 ', fg='white', bg='black',
                    command=lambda: press(0), height=2, width=8,font=('calbri 12 bold'))
    button0.grid(row=6, column=1)
 
    plus = Button(gui, text=' + ', fg='white', bg='black',
                command=lambda: press("+"), height=2, width=8,font=('calbri 12 bold'))
    plus.grid(row=5, column=3)
 
    minus = Button(gui, text=' - ', fg='white', bg='black',
                command=lambda: press("-"), height=2, width=8,font=('calbri 12 bold'))
    minus.grid(row=4, column=3)
 
    multiply = Button(gui, text=' × ', fg='white', bg='black',
                    command=lambda: press("*"), height=2, width=8,font=('calbri 12 bold'))
    multiply.grid(row=3, column=3)
 
    divide = Button(gui, text=' ÷ ', fg='white', bg='black',
                    command=lambda: press("/"), height=2, width=8,font=('calbri 12 bold'))
    divide.grid(row=2, column=3)
    modulus = Button(gui, text=' % ', fg='white', bg='black',
                    command=lambda: press("%"), height=2, width=8,font=('calbri 12 bold'))
    modulus.grid(row=2, column=2)
    raised_to = Button(gui, text=' ^ ', fg='white', bg='black',
                    command=lambda: press("**"), height=2, width=8,font=('calbri 12 bold'))
    raised_to.grid(row=2, column=1) 
    raised_to = Button(gui, text='//', fg='white', bg='black',
                    command=lambda: press("//"), height=2, width=8,font=('calbri 12 bold'))
    raised_to.grid(row=2, column=0) 
    square = Button(gui, text='x\u00b2', fg='white', bg='black',
                    command=lambda: press("**2"), height=2, width=8,font=('calbri 12 bold'))
    square.grid(row=1, column='2') 
    cube = Button(gui, text='x\u00b3', fg='white', bg='black',
                    command=lambda: press("**3"), height=2, width=8,font=('calbri 12 bold'))
    cube.grid(row=1, column='1')
    sqaure_root = Button(gui, text='\u221Ax', fg='white', bg='black',
                    command=lambda: press("**0.5"), height=2, width=8,font=('calbri 12 bold'))
    sqaure_root.grid(row=1, column='0')
    equal = Button(gui, text=' = ', fg='white', bg='black',
                command=equalpress, height=2, width=8,font=('calbri 12 bold'))
    equal.grid(row=6, column=3)
 
    clear = Button(gui, text='Clear', fg='white', bg='black',
                command=clear, height=2, width=8,font=('calbri 12 bold'))
    clear.grid(row=6, column='0')
    clearsingle = Button(gui, text='⌫', fg='white', bg='black',
                command=clearsingle, height=2, width=8,font=('calbri 12 bold'))
    clearsingle.grid(row=1, column='3')
 
    Decimal= Button(gui, text='.', fg='white', bg='black',
                    command=lambda: press('.'), height=2, width=8,font=('calbri 12 bold'))
    Decimal.grid(row=6, column=2)

    gui.mainloop()
```
